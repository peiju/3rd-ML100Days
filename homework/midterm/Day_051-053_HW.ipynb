```python
import pandas as pd
import numpy as np
import copy, time
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = 'data/'
app_train = pd.read_csv(data_path + 'train_data.csv')
app_test = pd.read_csv(data_path + 'test_features.csv')

train_Y = app_train['poi']
train_num = train_Y.shape[0]
ids = app_test['name']
df_train = app_train.drop(['name', 'poi'] , axis=1)
df_test = app_test.drop(['name'] , axis=1)
df = pd.concat([df_train,df_test])
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')

# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df.head()
```

# 缺值填補
* 補0: ['deferral_payments', 'loan_advances', 'bonus', 'restricted_stock_deferred', 'deferred_income', 'total_stock_value', 'exercised_stock_options', 'other', 'long_term_incentive', 'restricted_stock', 'director_fees', 'to_messages', 'from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi']
* 補眾數: ['salary', 'total_payments', 'expenses']

```python
import copy
df_fixed = copy.deepcopy(df)
df_fixed.head(10)
```

```python
def drop_col(df, col_name, cutoff=0.5):
    n = len(df)
    cnt = df[col_name].count()
    if (float(cnt) / n) < cutoff:
        df.drop(col_name, axis=1, inplace=True)
        print(col_name)
        
for feature in num_features:
    drop_col(df_fixed, feature, cutoff=0.5)
    
df_fixed.head()
```

```python
# 特徵工程
fin = ['salary', 'deferral_payments', 'total_payments', 'loan_advances', 'bonus', 'restricted_stock_deferred', 'deferred_income', 'total_stock_value', 'expenses', 'exercised_stock_options', 'other', 'long_term_incentive', 'restricted_stock', 'director_fees']
rm = ['deferral_payments', 'deferred_income', 'director_fees', 'loan_advances', 'long_term_incentive', 'restricted_stock_deferred']

for feature in num_features:
    if feature in rm:
        continue
    if feature in fin:
        df_fixed[feature] = np.log1p(df_fixed[feature].abs())
    else:
        q_75 = np.percentile(df_fixed[~df_fixed[feature].isnull()][feature], q = 90)
        df_fixed.loc[df_fixed[feature]>q_75, feature] = q_75        

# df_fixed = pd.DataFrame(MinMaxScaler((0, 1)).fit_transform(df_fixed), columns=df_fixed.columns)

df_fixed.head(10)
```

```python
from scipy.stats import mode

fill_zero = ['deferral_payments', 'loan_advances', 'bonus', 'restricted_stock_deferred', 'deferred_income', 'total_stock_value', 'exercised_stock_options', 'other', 'long_term_incentive', 'restricted_stock', 'director_fees', 'to_messages', 'from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi']
fill_mode = ['salary', 'total_payments', 'expenses']

for f in fill_zero:
    if f in rm:
        continue
    df_fixed.loc[df_fixed[f].isnull(), f] = 0
    
# for f in fill_mode:
#     mode_num = mode(df_fixed[~df_fixed[f].isnull()][f])[0]
#     print(f, mode_num)
#     df_fixed.loc[df_fixed[f].isnull(), f] = mode_num    
    
df_fixed.head(10)
```

```python
from fancyimpute import KNN, SoftImpute, IterativeImputer

#Use k=5 nearest rows which have a feature to fill in each row's missing
df_fixed = pd.DataFrame(KNN(k=5).fit_transform(df_fixed), columns=df_fixed.columns)
# df_fixed = pd.DataFrame(IterativeImputer().fit_transform(df_fixed), columns=df.columns)

df_fixed.head(10)
```

```python
email_features = ['to_messages', 'from_poi_to_this_person', 'from_messages', 'from_this_person_to_poi', 'shared_receipt_with_poi']
for feature in email_features:
    df_fixed[feature] = MinMaxScaler((0, 1)).fit_transform(df_fixed[[feature]])
```

```python
train_X = df_fixed[:train_num]
test_X = df_fixed[train_num:]
```

```python
# 特徵選擇
# SelectKBest 挑選出K個分數最高的特徵
from sklearn.feature_selection import SelectKBest
# 這是一稱為aNOVA的統計檢驗方法
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import chi2

k_best = SelectKBest(chi2, k=10)
train_X = k_best.fit_transform(train_X, train_Y)
print(train_X.shape)

test_X = k_best.transform(test_X)
print(test_X.shape)
```

```python
# 計算基礎分數
from sklearn.naive_bayes import GaussianNB
estimator = GaussianNB()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()

# model = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
# model.fit(train_X, train_Y)
# pred = model.predict_proba(test_X)[:, 1]
```

```python
from sklearn.neural_network import MLPClassifier

estimator = MLPClassifier(solver='lbfgs', activation='relu', random_state=1, verbose=False)
cross_val_score(estimator, train_X, train_Y, cv=5).mean()

# estimator.fit(train_X, train_Y)
# pred = estimator.predict_proba(test_X)[:, 1]
```
